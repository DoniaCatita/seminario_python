```python
def print_final_score(total_sc):
    final_score = [(player,score) for player, score in total_sc.items()]
    final_score.sort(key=lambda x: x[1]['points'], reverse = True)
    print(f'Ranking final:')
    print(f'jugador | kills | asistencias | muertes | MVP | puntos')
    for player, score in final_score:
        print(f'{player}{' '*8}{score['kills']}{' '*8}'+
        f'{score['assists']}{' '*10}{int(score['deaths'])}{' '*8}{score['mvp']}{' '*8}{score['points']}')

def act_total_score(total_sc,round_sc):
    for player, score in round_sc:
        for stat, value in score.items():
            total_sc[player][stat] = total_sc[player][stat] + value #al jugador player, actualizo la estadistica stat sumando el valor de la ronda
    

def print_round(round_sc,i):
    print(f'Ranking ronda {i}:')
    print(f'jugador | kills | asistencias | muertes | MVP | puntos')
    for player, score in round_sc:
        print(f'{player}{' '*8}{score['kills']}{' '*8}'+
        f'{score['assists']}{' '*10}{int(score['deaths'])}{' '*8}{score['mvp']}{' '*8}{score['points']}')
        

def get_round_score(scoreboard, mvp):
    # recibe el score de la ronda y calcula el puntaje total de cada jugador 
    round_score = [(player, score) for player, score in scoreboard.items()]
    for player, score in round_score:
        score['mvp'] = 0
        score['points'] = score['kills'] * 3 + score['assists'] + (-1) * score['deaths']
        if mvp['points'] == None or score['points'] > mvp['points']:
            mvp['points'] = score['points']
            mvp['user'] = player
    #    get_mvp(score['puntos'])
    return round_score




rounds = [
{
'Shadow': {'kills': 2, 'assists': 1, 'deaths': True},
'Blaze': {'kills': 1, 'assists': 0, 'deaths': False},
'Viper': {'kills': 1, 'assists': 2, 'deaths': True},
'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 0, 'assists': 2, 'deaths': False},
'Blaze': {'kills': 2, 'assists': 0, 'deaths': True},
'Viper': {'kills': 1, 'assists': 1, 'deaths': False},
'Frost': {'kills': 2, 'assists': 1, 'deaths': True},
'Reaper': {'kills': 0, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 1, 'assists': 0, 'deaths': False},
'Blaze': {'kills': 2, 'assists': 2, 'deaths': True},
'Viper': {'kills': 1, 'assists': 1, 'deaths': True},
'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 2, 'assists': 1, 'deaths': False},
'Blaze': {'kills': 1, 'assists': 0, 'deaths': True},
'Viper': {'kills': 0, 'assists': 2, 'deaths': False},
'Frost': {'kills': 1, 'assists': 1, 'deaths': True},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 1, 'assists': 2, 'deaths': True},
'Blaze': {'kills': 0, 'assists': 1, 'deaths': False},
'Viper': {'kills': 2, 'assists': 0, 'deaths': True},
'Frost': {'kills': 1, 'assists': 1, 'deaths': False},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': True}
}
]

i = 0
total_sc = {player: {'kills': 0, 'assists': 0, 'deaths': 0, 'mvp': 0, 'points': 0} for player in rounds[0].keys()}
for round in rounds:
    i = i + 1
    mvp = {'points': None}
    round_sc = get_round_score(round,mvp) # genero lista de tuplas (nombre,estadisticas_jugador)
    round_sc.sort(key=lambda x: x[1]['points'], reverse = True) #ordeno la lista segun puntaje total
    round_sc[0][1]['mvp'] = 1 #marco al mvp (lista[1eratupla][estadisticas][mvp] = 1)
    print_round(round_sc,i) 
    print()
    act_total_score(total_sc,round_sc) #actualizo el puntaje total con los datos de la ronda actual
print_final_score(total_sc)  
```
